```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
# 1. Simulate an Embedding Model (for demonstration)
def get_embedding(text):
  # Real embeddings capture semantic meaning.
  seed = sum(ord(c) for c in text) % 1000
  np.random.seed(seed)
  return np.random.rand(10) # 10-dimensional vector
```

```python
# 2. Prepare our 'memory' (documents/conversation snippets)
memory_documents = [
    "The user is interested in simple chatbots.",
    "The chatbot's name is Chatbot.",
    "The conversation included a joke about atoms.",
    "The user asked how to save conversation to a file.",
    "The user wants to append new entries to the conversation file.",
    "The current topic is upgrading to agentic AI."
    ]
```

```python
# 3. Create our 'Vector Database’
vector_db = []
for doc in memory_documents:
  embedding = get_embedding(doc)
  vector_db.append({'text': doc, 'embedding': embedding})
print(f"Created a simple in-memory vector store with {len(vector_db)} entries.")
```

```python
def retrieve_similar_memories(query_text, top_k=2):
  query_embedding = get_embedding(query_text)
  similarities = []
  for entry in vector_db:
    doc_embedding = entry['embedding'].reshape(1, -1)
    query_embedding_reshaped = query_embedding.reshape(1, -1)
    similarity = cosine_similarity(query_embedding_reshaped, doc_embedding)[0][0]
    similarities.append((entry['text'], similarity))
  similarities.sort(key=lambda x: x[1], reverse=True)
  return similarities[:top_k]
```

```python
print("\n--- Demonstrating Memory Retrieval ---")
query = "What did we talk about saving files?"
retrieved_memories = retrieve_similar_memories(query)
print(f"Query: '{query}'")
print("Top 2 relevant memories:")
for text, score in retrieved_memories:
  print(f"- '{text}' (Similarity: {score:.4f})")
```

```python
query = "Tell me about the chatbot's identity."
retrieved_memories = retrieve_similar_memories(query)
print(f"\nQuery: '{query}'")
print("Top 2 relevant memories:")
for text, score in retrieved_memories:
  print(f"- '{text}' (Similarity: {score:.4f})")
```
