## **Simple Rule-Based Chatbot Example **

```python
def simple_chatbot():
  print("Hello! I am a simple chatbot. Type 'quit' to exit.")
  while True:
    user_input = input("You: ").lower()
    if user_input == 'quit':
      print("Chatbot: Goodbye!")
      break
    elif 'hello' in user_input or 'hi' in user_input:
      print("Chatbot: Hello! How can I assist you today?")
    elif 'how are you' in user_input:
      print("Chatbot: I'm just a chatbot, but I'm doing hreat and here to help you! Thanks for asking.")
    elif 'name' in user_input:
      print("Chatbot: I'm just a chatbot, so I don't have a name.")
    elif 'weather' in user_input:
      print("Chatbot: I'm sorry, I can't provide real-time weather information. You can check a weather website for that.")
    elif 'help' in user_input:
      print("Chatbot: I can answer basic questions like 'hello', 'how are you', 'name', 'weather', and 'help'.")
    elif 'creator' in user_input:
      print("Chatbot: I was created by Akanksha.")
    else:
      print("Chatbot: I'm sorry, I didn't understand that. Can you please rephrase?")
```

```python
simple_chatbot()
```

```python
# Simulate a chat conversation history
conversation_history = [
    "Chatbot: Hello! I'm a simple chatbot. Type 'quit' to exit.",
    "You: hello",
    "Chatbot: Hi there! How can I help you today?",
    "You: how are you",
    "Chatbot: I'm just a program, but I'm doing great! Thanks for asking.",
    "You: what is your name",
    "Chatbot: I don't have a name. You can call me Chatbot!",
    "You: quit", "Chatbot: Goodbye!"
    ]
```

```python
filename = "chatbot_conversation.txt"
try:
  with open(filename, "w") as f:
    for line in conversation_history:
      f.write(line + "\n")
  print(f"Conversation successfully saved to {filename}")
except IOError as e:
  print(f"Error saving conversation to file: {e}")
```

```python
try:
  with open(filename, "r") as f:
    content = f.read()
  print("\n--- Contents of " + filename + " ---\n")
  print(content)
  print("\n----------------------------------")
except FileNotFoundError:
  print(f"Error: The file {filename} was not found.")
except IOError as e:
  print(f"Error reading file: {e}")
```
