### ***Agentic AI using JSON Dataset***

```python
class MyAgent:
  def __init__(self, name):
    self.name = name
    self.state = "idle"
    self.message_log = []
  def perceive(self, environment_data):
    #Agent's perception logic
    print(f"{self.name} perceiving: {environment_data}")
    if "alert" in environment_data:
      self.state = "alerted"
  def act(self):
    #Agent's action logic based on its state
    if self.state == "alerted":
      print(f"{self.name} acting Responding to alert.")
      self.state = "idle" #Reset state after acting
    else:
      print(f"{self.name} acting Performing routine tasks.")
  def communicate(self, recipient_agent, message):
    log_entry = {'type': 'sent', 'to':'recipient_agent.name', 'message': message}
    self.message_log.append(log_entry) #log sent message
    print(f"{self.name} sending message to {recipient_agent.name}: {message}")
    recipient_agent.message_log.append({'type': 'received', 'from': self.name, 'message': message})
  def receive_message(self, sender_name, message):
    log_entry = {'type': 'received', 'from': sender_name, 'message': message}
    self.message_log.append(log_entry) #log received message
    print(f"{self.name} receiving message from {sender_name}: {message}")
  def display_log(self):
    print(f"{self.name}'s message log ---")
    if not self.message_log:
      print("No messages in log.")
    for entry in self.message_log:
      if entry['type'] == 'sent':
        print(f"Sent to {entry['to']}: '{entry['message']}'")
      elif entry['type'] == 'received':
        print(f"Received from {entry['from']}: '{entry['message']}'")
```

```python
# Example usage with communication and logging:
agent1 = MyAgent("Alpha")
agent2 = MyAgent("Beta")

print("---Communication example with logging---")
agent1.communicate(agent2,"Hello Beta,how are things?")
agent2.communicate(agent1,"All good Alpha,just monitoring.")
agent1.display_log()
agent2.display_log()
```

```python
import json

def save_agent_log_to_json(agent,filename=None):
  if filename is None:
    filename = f"{agent.name.lower()}_message_log.json"
  with open(filename, 'w') as f:
    json.dump(agent.message_log, f, indent=4)
  print(f"Log for {agent.name} saved to {filename}")
def load_agent_log_from_json(agent,filename=None):
  if filename is None:
    filename = f"{agent.name.lower()}_message_log.json"
  try:
    with open(filename, 'r') as f:
      agent.message_log = json.load(f)
    print(f"Log for {agent.name} loaded from {filename}")
  except FileNotFoundError:
    print(f"No log file found for {agent.name} at {filename}. Initializing empty log.")
    agent.message_log = []
  except json.JSONDecodeError:
    print(f"Error decoding JSON from {filename}. Initializing empty log.")
    agent.message_log = []
```

```python
agent1_json = MyAgent("Alpha.json")
agent2_json = MyAgent("Beta.json")

print("---Communication example with JSON logging---")
agent1_json.communicate(agent2_json,"Hello Beta.json,can you hear me?")
agent2_json.communicate(agent1_json,"Great! Saving our conversation.")
agent1_json.display_log()
agent2_json.display_log()
```

```python
save_agent_log_to_json(agent1_json)
save_agent_log_to_json(agent2_json)

#Demonstrate loading: Create new agents and load their logs
print("\n---Demonstrating log loading---")
agent3_loaded = MyAgent("Alpha.json") #will load Alpha.json's log
agent4_loaded = MyAgent("Beta.json") #will load Beta.json's log

load_agent_log_from_json(agent3_loaded)
load_agent_log_from_json(agent4_loaded)

agent3_loaded.display_log()
agent4_loaded.display_log()
```
